- Load into a dataframe NBA data from github url: https://raw.githubusercontent.com/fivethirtyeight/data/master/nba-draymond/draymond.csv 
- show the first lines

```python
import requests
import pandas as pd

# Insert your code here
download_url = "https://raw.githubusercontent.com/fivethirtyeight/data/master/nba-draymond/draymond.csv"
target_csv_path = "data/draymond.csv"

response = requests.get(download_url)
response.raise_for_status()    
with open(target_csv_path, "wb") as f:
    f.write(response.content)
players = pd.read_csv(target_csv_path)
players.head()
```

- Load into a dataframe the tattoo data from NBA players from github url: https://raw.githubusercontent.com/fivethirtyeight/data/master/nba-tattoos/nba-tattoos-data.csv"
- Show the first lines  

```python
# Insert your code here
download_url = "https://raw.githubusercontent.com/fivethirtyeight/data/master/nba-tattoos/nba-tattoos-data.csv"
target_csv_path = "tattoos.csv"

response = requests.get(download_url)
response.raise_for_status()    
with open(target_csv_path, "wb") as f:
    f.write(response.content)
tattoos = pd.read_csv("tattoos.csv")
tattoos.head()
```

Merge data: put toggeter the information from each dataset for each player

```python
# Insert your code here
pd.merge(players, tattoos, left_on='player', right_on='Player Name', how='inner')
```
